```python
import pandas as pd
import numpy as np
import datetime as dt

import itertools
import json

from pprint import pprint
```

```python
# load up some web traffic tick data
ars_data = json.load(open("rtdata/arstechnica.json"))
slate_data = json.load(open("rtdata/slate.json"))
```

```python
# take a peek inside
ars_data.keys()
```

```python
ars_data["timeline"][0:10]
```

```python
top_posts = ars_data["top_keys"][0:2]
for post in top_posts:
    pprint(dict(info=post["info"], views=post["views"], total_24h=int(sum(post["historical"]))))
```

```python
# let's start with the site timelines

# the index is the same, a series of datetime ticks, 
# but in string format -- so we'll convert these
index = [pd.to_datetime(date)
            for date, datum in ars_data["timeline"]]
# the ticks themselves are floats, which are fine
ars_ticks = [datum for date, datum in ars_data["timeline"]]
slate_ticks = [datum for date, datum in slate_data["timeline"]]
# a dataframe is just some tick data and an index
df1 = pd.DataFrame(ars_ticks, index=index)
df1.plot()
df2 = pd.DataFrame(slate_ticks, index=index)
df2.plot()
```

```python
# pandas is clever, you can easily add two DataFrames with aligned indexes
(df1 + df2).plot(legend=False)
```

```python
# or, 12-hour period from 10/16 @ 9pm to 10/17 @ 9am
df1["2013-10-16T21:00":"2013-10-17T9:00"].plot(legend=False)
```

```python
# it's also pretty easy to merge these DataFrames,
# so that both timelinesare in a single data frame
# exposed as labeled columns
df = df1.copy()
df["arstechnica.com"] = df1[0]
df["slate.com"] = df2[0]
del df[0]
df
```

```python
# now, plotting this dataframe just does the right thing
df.plot()
```

```python
# what were the high points?
df.max()
```

```python
# and totals?
df.sum().plot(kind="barh")
```

```python
# at what times?
df.idxmax()
```

```python
# when was the peak 1-hour period?
df.resample("60Min", how="mean").idxmax()
```

```python
# What was resample? 
# With high-granularity data, Pandas can "resample" 
# ticks to smooth out our results
df.resample("60Min").plot()
```

```python
# this is flexible to point of absurdity
df.resample("1h15min", how="max").plot()
```

```python
# let's use some of our other tick data -- this time on posts
posts = {}
for post in itertools.chain(ars_data["top_keys"], slate_data["top_keys"]):
    posts[post["info"]["title"].encode("utf-8")[0:40]] = post["historical"]
post_df = pd.DataFrame(posts, index=index)
post_df
```

```python
# this will be noisy, but show convenience of plot()
post_df.plot(legend=False)
```

```python
# subplots will allow us to do a vertical sparkline-like compare of all post data
post_df.plot(subplots=True, figsize=(15, 30), legend=False, grid=False, mark_right=False)
```

```python
# some magic to initialize vincent with IPython Notebook
import vincent
vincent.core.initialize_notebook()

d3_js_url = "/files/js/d3.v3.min.js"
vega_js_url = '/files/js/vega.js'

def initialize_notebook():
    """Initialize the iPython notebook display elements"""
    try:
        from IPython.core.display import display, HTML, Javascript
    except ImportError:
        print('iPython Notebook could not be loaded.')

    display(HTML('<script src="%s"></script>' % d3_js_url))
    display(HTML('<script src="%s"></script>' % vega_js_url))

initialize_notebook()
```

```python
# plug the post dataframe into a Vincent StackedArea graph
post_stack = vincent.StackedArea(post_df)
post_stack.axis_titles(x='Date', y='Pageviews')
post_stack.legend(title='Pulse posts')
post_stack.display()
```
