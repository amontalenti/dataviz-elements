```python
conn = pymongo.MongoClient()
coll = conn["parsely_insights"]["refdomain_destinations"]
coll.count()
```

```python
# different kinds of "top URLs" for different referrers
fields = {"top_destinations": 1, "date": 1, "referring_domain": 1}
criteria = {"date": {"$gte": last_run-dt.timedelta(days=90), "$lt": last_run}}
criteria.update({"referring_domain": {"$in": ["facebook.com", "t.co", "pinterest.com", "plus.google.com", "linkedin.com"]}})
social_urls = coll.find(criteria, fields)
criteria.update({"referring_domain": {"$in": ["google.com", "bing.com", "yahoo.com"]}})
searched_urls = coll.find(criteria, fields)
criteria.update({"referring_domain": {"$in": ["drudgereport.com", "digg.com", "reddit.com", "realclearpolitics.com"]}})
aggregated_urls = coll.find(criteria, fields)
criteria.update({"referring_domain": {"$in": ["feedly.com"]}})
rss_urls = coll.find(criteria, fields)
```

```python
MIN_DAILY_PVS = 10000
agg = coll.find({
                "hourly_pvs.total": {"$gt": MIN_DAILY_PVS},
                "date": {"$gte": last_run-dt.timedelta(days=90), "$lt": last_run}
                }, {"referring_domain": 1, "_id": 0, "hourly_pvs": 1, "date": 1})
hourlies = []
for record in agg:
    for hour, value in record["hourly_pvs"].iteritems():
        if hour == "total": continue
        date = record["date"]
        date = date.replace(hour=int(hour))
        new_record = {
                   "domain": record["referring_domain"],
                   "date": date,
                   "pvs": value
        }
        hourlies.append(new_record)
```

```python
raw = pd.DataFrame(list(hourlies))
df = raw.set_index("date")
```

```python
df[df.domain == "__alldomains__"].pvs.resample("7D").plot()
```

```python
def plot_domains(*domains):
    for domain in domains:
        df[df.domain == domain].pvs.resample("7D").plot()
```

```python
# in last 30 months, facebook is growing as 
# a source to news sites;
# twitter and pinterest are relatively flat
plot_domains("facebook.com", "t.co", "pinterest.com")
```

```python
# Google and Facebook growing at a similar pace, despite
# that Google remains 2X the size of Facebook as a traffic driver
plot_domains("google.com", "facebook.com")
```

```python
# editorial front-page, crowd front-page & personalized front-page
# the changing nature of news consumer "home-pages"
plot_domains("drudgereport.com", "reddit.com", "feedly.com")
```

```python
# bing and yahoo are still fighting for the scraps of search
# but: importantly, these are the only two sources where you
# can still get keyword data!
plot_domains("bing.com", "ask.com", "duckduckgo.com")
```
